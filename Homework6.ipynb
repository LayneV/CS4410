```python
%matplotlib inline
import nltk

nltk.download('stopwords')

nltk.download('wordnet')

nltk.download('punkt')
```

```python
!pip install requests

import requests

target_url ='http://www.gutenberg.org/files/2265/2265.txt'

response = requests.get(target_url)

data = response.text
```

```python
!pip install textblob
import textblob
from textblob import TextBlob

blob=TextBlob(data)
```

```python
import imageio

image_file="https://media.cheggcdn.com/media/216/21621ee5-e80f-47f3-9145-513f2229b390/phploeBuh.png"

mask_image = imageio.imread(image_file)
```

```python
items = blob.word_counts.items()
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
items = [item for item in items if item[0] not in stop_words]
```

```python
from operator import itemgetter
sorted_items = sorted(items, key=itemgetter(1), reverse=True)
```

```python
top20 = sorted_items[:20]
```

```python
import pandas as pd
df = pd.DataFrame(top20, columns=['word', 'count'])
```

```python
df
```

```python
axes = df.plot.bar(x='word', y='count', legend=False)

import matplotlib.pyplot as plt

plt.gcf().tight_layout()
```

```python
from wordcloud import WordCloud

word_freq_dict = dict(items)

wordcloud = WordCloud(width=1000, height=1000,
    colormap='prism', mask=mask_image, background_color='white').generate_from_frequencies(word_freq_dict)

wordcloud.to_file('Hamlet.png')

```
